```python
#W9A2

import heapq

def kth_largest(matrix, k):
    n = len(matrix)
    m = len(matrix[0])

    heap = []
    visited = set()

    heapq.heappush(heap, (-matrix[n-1][m-1], n-1, m-1))
    visited.add((n-1, m-1))

    for _ in range(k - 1):
        val, i, j = heapq.heappop(heap)

        if i - 1 >= 0 and (i-1, j) not in visited:
            heapq.heappush(heap, (-matrix[i-1][j], i-1, j))
            visited.add((i-1, j))

        if j - 1 >= 0 and (i, j-1) not in visited:
            heapq.heappush(heap, (-matrix[i][j-1], i, j-1))
            visited.add((i, j-1))

    return -heapq.heappop(heap)[0]
```

```python
#W9A3

def min_after_one_swap(num):
    s = list(str(num))
    n = len(s)
    ans = None

    for i in range(n):
        for j in range(i + 1, n):
            s[i], s[j] = s[j], s[i]

            if s[0] != '0':
                val = int("".join(s))
                if ans is None or val < ans:
                    ans = val

            s[i], s[j] = s[j], s[i]

    return ans
```

```python
#W9A8

def is_path_crossing(moves):
    x, y = 0, 0
    visited = set()
    visited.add((0, 0)) 

    for move in moves:
        if move == 'R':
            x += 1
        elif move == 'L':
            x -= 1
        elif move == 'U':
            y += 1
        elif move == 'D':
            y -= 1

        if (x, y) in visited:
            return True

        visited.add((x, y))

    return False
```

```python
#W9A9

def Median(nums):
    a, b, c, d, e = nums

    if a > b:
        a, b = b, a
    if c > d:
        c, d = d, c
    if a > c:
        a, c = c, a
        b, d = d, b
    if b > e:
        b, e = e, b
    if b > c:
        b, c = c, b
    if b > d:
        b, d = d, b

    return b
```

```python
#W9A10

def findRadius(houses, heaters):
    houses.sort()
    heaters.sort()

    def can_cover(r):
        j = 0  
        for h in houses:
            while j < len(heaters) and heaters[j] + r < h:
                j += 1
            if j == len(heaters) or heaters[j] - r > h:
                return False
        return True

    left, right = 0, 10**9
    ans = right

    while left <= right:
        mid = (left + right) // 2
        if can_cover(mid):
            ans = mid
            right = mid - 1
        else:
            left = mid + 1

    return ans
```
