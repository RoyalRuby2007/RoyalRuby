```python
def PrimeF(n):
    num = [1] * (n + 1)
    F = [0] * (n + 1)
    S = int(n ** 0.5)
    for i in range(2 , S + 1):
        if (num[i] == 1):
            for j in range(i * i , n + 1 , i):
                num[j] = 0
    num[0] = num[1] = 0
    for s in range(2 , n + 1):
        if (num[s] == 1):
            F[s] = F[s - 1] + s
        else:
            F[s] = F[s - 1]
    return F

F = PrimeF(1000000)

ans = []
while (True):
    try:
        a , b = map(int , input().split())
        ans.append(F[b] - F[a - 1])
    except:
        break;

for k in ans:
    print (k)
```

```python
n = int(input())

def Palindrome(n):
    n = str(n)
    Length = len(n)
    for i in range(Length // 2 + 1):
        if (n[i] != n[Length - 1 - i]):
            return False
    return True

ans = 0
while (Palindrome(n) == False):
    n = n + int(str(n)[::-1])
    ans += 1

print (ans , n)
```

```python
n = int(input())

def check(n):
    n = str(n)
    A = [k for k in n]
    A = set(A)
    return (len(n) == len(A))

ans = []
for i in range(1 , n + 1):
    if ((i ** 0.5 == (int(i ** 0.5))) and check(i)):
        ans.append(i)

if (len(ans) == 0):
    print ("no number")
else:
    for k in ans:
        print (k , end = " ")
```

```python
n = int(input())

Dt = {1 : 1}
def Collatz(n):
    if (Dt.get(n) == None):
        Dt[n] = 1
    if (Dt.get(n // 2) == None):
            Dt[n // 2] = 1
    if (Dt.get(n * 3 + 1) == None):
            Dt[n * 3 + 1] = 1
    if (n == 1 or Dt[n] > 1):
        return
    if (n % 2 == 0):
        if (n // 2 == 1 or Dt[n // 2] > 1):
            Dt[n] += Dt[n // 2]
            return
        else:
            Collatz(n // 2)
            Dt[n] += Dt[n // 2]
            return
    else:
        if (Dt[n * 3 + 1] > 1):
            Dt[n] += Dt[n * 3 + 1]
            return
        else:
            Collatz(n * 3 + 1)
            Dt[n] += Dt[n * 3 + 1]
            return

for i in range(1 , n + 1):
    Collatz(i)

ans = 1
Max = 1
for k in Dt:
    if (Dt[k] > Max):
        ans = k
        Max = Dt[k]

print (ans , Max)
```

```python
n = int(input())

def phantich(n):
    if (n % 2 == 0):
        Dt = {}
        Count = 0
        i = 2
        while (i * i <= n):
            while (n % i == 0):
                if (Dt.get(i) == None):
                    Dt[i] = 1
                else:
                    Dt[i] += 1
                if (i % 2 != 0):
                    Count += 1
                n //= i
            i += 1
        if (n > 1):
            if (Dt.get(i) == None):
                Dt[i] = 1
            else:
                Dt[i] += 1
            if (n % 2 != 0):
                Count += 1
        ans = 1
        for k in Dt:
            ans *= (Dt[k] + 1)
        ans -= (Count * (Count + 1)) // 2 + 1
    else:
        ans = 0
    return ans

print (phantich(n))
```
