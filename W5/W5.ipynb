```python
#W5A3

n = int(input())

def Songuyento(n):
    if (n < 2):
        return False
    if (n == 2):
        return True
    if (n % 2 == 0):
        return False
    S = n ** 0.5
    for i in range(3 , S + 1 , 2):
        if (n % i == 0):
            return False
    return True

print (Songuyento(n))
```

```python
#W5A4

n = int(input())

def phantich(n):
    Dt = {}
    i = 2
    while (i * i <= n):
        while (n % i == 0):
            if (Dt.get(i) == None):
                Dt[i] = 1
            else:
                Dt[i] += 1
            n //= i
        i += 1
    if (n > 1):
        if (Dt.get(n) == None):
            Dt[n] = 1
        else:
            Dt[n] += 1
    return Dt

def PerfectNum(n):
    if (n > 0):
        Dt = phantich(n)
        ans = 1
        for k in Dt:
            ans *= ((k ** (Dt[k] + 1) - 1) // (k - 1))
        return (ans == 2 * n)
    return False

print (PerfectNum(n))        
```

```python
#W5A8

a , b = map(int , input().split())

def Solve(a , b):
    x = bin(a ^ b)
    if (x[0] == "-"):
        ans = sum(list(map(int , [v for v in x[3 :]])))
    else:
        ans = sum(list(map(int , [v for v in x[2 :]])))
    return ans

print (Solve(a , b))
```

```python
#W5A9

n = int(input())

def SumDNum(n):
    ans = sum(list(map(int , [x for x in str(n)])))
    return ans

print (SumDNum(n))
```

```python
#W5A10

S1 , S2 = (input().split())

def Solve(S1 , S2):
    Dt1 = {}
    Dt2 = {}
    n = len(S1)
    for i in range(n):
        if (Dt1.get(S1[i]) == None):
            Dt1[S1[i]] = S2[i]
        else:
            if (Dt1[S1[i]] != S2[i]):
                return False
            
        if (Dt2.get(S2[i]) == None):
            Dt2[S2[i]] = S1[i]
        else:
            if (Dt2[S2[i]] != S1[i]):
                return False
    return True

print (Solve(S1 , S2))
```
