```python
#W6A4

D = list(map(str , input().split()))

Dt = {}
for k in D:
    n = len(k)
    i = k.index(":")
    if (Dt.get(k[:i]) == None):
        Dt[k[:i]] = [(str(k[i + 1 :]))]
    else:
        Dt[k[:i]] = Dt[k[:i]] + [(str(k[i + 1 :]))]
print (Dt)
```

```python
#W6A9

S1 = input()
S2 = input()

S1 = S1[1 : - 1]
S2 = S2[1 : - 1]
Dt = {}
D1 = []
D2 = []
if (len(S1) != 0):
    D1 = list(map(str , S1.split(",")))
if (len(S2) != 0):
    D2 = list(map(str , S2.split(",")))
D = (D1 + D2)
for k in D:
    k = k.replace(" " , "")
    x , y = k.split(":")
    x = x[1 : -1]
    if (Dt.get(x) == None):
        Dt[x] = int(y)
    else:
        Dt[x] += int(y)
Dt = dict(sorted(Dt.items()))
print (Dt)
    
```

```python
#W6A10

D = list(map(int , input().split()))
k = int(input())

D.reverse()
Dt = {}
Ans = []
for x in D:
    if (Dt.get(k - x) == None):
        Dt[x] = 1
    else:
        Ans.append((x , k - x))
        Dt[x] = 1
Ans = sorted(list(set(Ans)))
print (Ans)
```

```python
#W6A12

D = list(map(int , input().split()))

Dt = {}
Max = 1
for k in D:
    if (Dt.get(k) == None):
        Dt[k] = 1
    else:
        Dt[k] += 1
        Max = max(Max , Dt[k])
Dt = dict(sorted(list(Dt.items())))
for k in Dt:
    if (Dt[k] == Max):
        print (k)
        break
```

```python
#W6A14

A = list(map(int , input().split()))
B = list(map(int , input().split()))

Dt = {}
for k in A:
    if (Dt.get(k) == None):
        Dt[k] = 1
for k in B:
    if (Dt.get(k) != None):
        Dt[k] += 1
D = []
for k in Dt:
    if (Dt[k] > 1):
        D.append(k)
print (D)
```

```python
#W6A21

D = list(input().split())

Dt = {}
n = len(D)
for i in range(n):
    if (Dt.get(D[i]) == None):
        Dt[D[i]] = tuple([i])
    else:
        ans = (*Dt[D[i]] , *tuple([i]))
        Dt[D[i]] = ans
print (Dt)
```
