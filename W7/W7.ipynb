```python
#W7A1

arr = list(map(int , input().split()))
target = int(input())

def binary_search(arr , left , right , target):
    mid = (left + right) // 2
    while (left <= right):
        if (target < arr[mid]):
            right = mid - 1
            mid = (left + right) // 2
        elif (target > arr[mid]):
            left = mid + 1
            mid = (left + right) // 2
        else:
            return mid
    return -1

print (binary_search(arr , 0 , len(arr) - 1 , target))
```

```python
#W7A2

arr = list(map(int , input().split()))
x = int(input())

def count_occurrences(arr, x) -> int:
    count = 0
    for k in arr:
        if (x == k):
            count += 1
    return count

print (count_occurrences(arr , x))
```

```python
#W7A9

from ast import literal_eval

num = literal_eval(input())
n = len(num)
num = [0] + num

BIT = [0] * 100001

def Update(i , x):
    while (i < 100001):
        BIT[i] += x
        i += (i) & (-i)

def Build():
    for i in range(1 , n + 1):
        Update(i , 0)

def Sum(N):
    i = N
    ans = 0
    while (i > 0):
        ans += BIT[i]
        i -= (i) & (-i)
    return ans

A = [0] * n
for i in range(n , 0 , -1):
    A[i - 1] = Sum(num[i] - 1)
    Update(num[i] , 1)
print (A)
```

```python
#W7A10

from ast import literal_eval
from bisect import bisect_right
from bisect import insort

nums = literal_eval(input())
m = int(input())

S = 0
ans = 0
F = [0]
for k in nums:
    S = (S + k) % m
    i = bisect_right(F , S)
    if (i < len(F)):
        ans = max(ans , (S - F[i] + m) % m)
    else:
        ans = max(ans , S)
    insort(F , S)
print (ans)
```

```python
#W7A13

n = int(input())
D = list(map(int , input().split()))

D.sort()
i = 0
j = 0
ans = 0
while (0 <= i <= j < n):
    if (D[j] - D[i] <= 1):
        ans = max(ans , j - i + 1)
        j += 1
    else:
        i += 1
print (ans)
```

```python
#W7A14

from bisect import bisect_right

n = int(input())
D = list(map(int , input().split()))

A = [(x , i) for i , x in enumerate(D)]
A.sort(key= lambda x: x[0])

idx = bisect_right(A , (7 , 10000000))
ans = []
if (idx < len(A) - 1):
    idx -= 1
    while (idx >= 0 and A[idx][0] == 7):
        ans.append(A[idx][1])
        idx -= 1
    ans.sort(reverse= True)
    for k in ans:
        print (k , end = " ")
else:
    print ("Not found") 
```
