```python
#W8A1

pi = 3.14

class Cylinder:
    def __init__(self , r , h):
        self.r = r
        self.h = h
    def S(self):
        ans = (pi * (self.r ** 2) * 2 + 2 * pi * self.r * self.h)
        return f"{ans:.2f}"
    def V(self):
        ans = (pi * (self.r ** 2) * self.h)
        return f"{ans:.2f}"
r , h = map(float , input().split())
Program = Cylinder(r , h)

print (Program.S())
print (Program.V())
```

```python
#W8A2

class Date:
    def __init__(self, day, month, year):
        self.day = day
        self.month = month
        self.year = year

    def is_leap_year(self):
        return (self.year % 400 == 0) or (self.year % 4 == 0 and self.year % 100 != 0)

    def get_days_in_month(self):
        days_per_month = [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
        
        if self.month == 2 and self.is_leap_year():
            return 29
        
        if self.month < 1 or self.month > 12:
            return 0
            
        return days_per_month[self.month]

    def is_valid(self):
        if self.year < 1: return False
        if self.month < 1 or self.month > 12: return False
        
        max_days = self.get_days_in_month()
        if self.day < 1 or self.day > max_days: return False
        
        return True

    def get_next_day(self):
        n_day = self.day
        n_month = self.month
        n_year = self.year

        n_day += 1

        if n_day > self.get_days_in_month():
            n_day = 1
            n_month += 1
            
            if n_month > 12:
                n_month = 1
                n_year += 1
        
        return f"{n_day:02d}/{n_month:02d}/{n_year}"

if __name__ == "__main__":
    try:
        input_str = input("Nhập ngày (dd/mm/yyyy): ")
        
        parts = input_str.split('/')
        
        if len(parts) != 3:
            print("INVALID")
        else:
            d, m, y = map(int, parts)
            
            my_date = Date(d, m, y)

            if my_date.is_valid():
                print(my_date.get_next_day())
            else:
                print("INVALID")
                
    except ValueError:
        print("INVALID")
```

```python
#W8A4

class Calculator:
    def __init__(self, a, b):
        self.a = a
        self.b = b

    def set_numbers(self, a, b):
        self.a = a
        self.b = b
        print(f"Đã cập nhật hai số mới: {self.a} và {self.b}")

    def add(self):
        return self.a + self.b

    def subtract(self):
        return self.a - self.b

    def product(self):
        return self.a * self.b

    def divide(self):
        if self.b == 0:
            return "Lỗi: Không thể chia cho 0"
        return self.a / self.b

    def power(self):
        return self.a ** self.b

    def mod(self):
        if self.b == 0:
            return "Lỗi: Không thể chia lấy dư cho 0"
        return self.a % self.b

def main():
    try:
        print("--- KHỞI TẠO MÁY TÍNH ---")
        num1 = float(input("Nhập số thứ nhất: "))
        num2 = float(input("Nhập số thứ hai: "))
        
        my_calc = Calculator(num1, num2)

        while True:
            print("\n============ MENU ============")
            print(f"Hai số hiện tại: {my_calc.a} và {my_calc.b}")
            print("1. Cộng (add)")
            print("2. Trừ (subtract)")
            print("3. Nhân (product)")
            print("4. Chia (divide)")
            print("5. Lũy thừa (power)")
            print("6. Chia lấy dư (mod)")
            print("7. Nhập lại 2 số mới (set_numbers)")
            print("==============================")

            choice = input("Chọn phép tính (1-7): ")

            if choice == '1':
                print(f"Kết quả: {my_calc.add()}")
            elif choice == '2':
                print(f"Kết quả: {my_calc.subtract()}")
            elif choice == '3':
                print(f"Kết quả: {my_calc.product()}")
            elif choice == '4':
                print(f"Kết quả: {my_calc.divide()}")
            elif choice == '5':
                print(f"Kết quả: {my_calc.power()}")
            elif choice == '6':
                print(f"Kết quả: {my_calc.mod()}")
            elif choice == '7':
                n1 = float(input("Nhập số thứ nhất mới: "))
                n2 = float(input("Nhập số thứ hai mới: "))
                my_calc.set_numbers(n1, n2)
            else:
                print("Lựa chọn không hợp lệ!")

            exit_ans = input("\nBạn có muốn thoát? (Gõ 'yes' để thoát): ")
            if exit_ans in ["yes", "YES", "Yes"]:
                print("Đã thoát chương trình.")
                break

    except ValueError:
        print("Lỗi: Vui lòng nhập số hợp lệ.")

if __name__ == "__main__":
    main()
```

```python
#W8A5

class ShoppingCart:
    def __init__(self):
        self.products = []

    def add_product(self, name, price):
        item = {"name": name, "price": price}
        self.products.append(item)
        print(f"Đã thêm '{name}' với giá {price} vào giỏ.")

    def remove_product(self, name):
        for i, item in enumerate(self.products):
            if item["name"] == name:
                removed_item = self.products.pop(i)
                print(f"Đã xóa sản phẩm '{removed_item['name']}' khỏi giỏ hàng.")
                return
        print(f"Lỗi: Không tìm thấy sản phẩm tên '{name}' trong giỏ.")

    def check_empty(self):
        if len(self.products) == 0:
            print("Giỏ hàng hiện đang TRỐNG.")
            return True
        else:
            print("Giỏ hàng KHÔNG trống.")
            return False

    def get_total(self):
        total = sum(item["price"] for item in self.products)
        print(f"Tổng tiền giỏ hàng: {total}")
        return total

    def show_cart(self):
        if len(self.products) == 0:
            print("Giỏ hàng trống!")
        else:
            print("--- Danh sách sản phẩm ---")
            for idx, item in enumerate(self.products, 1):
                print(f"{idx}. {item['name']} - Giá: {item['price']}")
            print("--------------------------")

    def clear_cart(self):
        self.products = []
        print("Đã xóa sạch giỏ hàng.")

def main():
    cart = ShoppingCart()

    while True:
        print("\n========== MENU ==========")
        print("2. Thêm sản phẩm")
        print("3. Xóa sản phẩm (theo tên)")
        print("4. Kiểm tra giỏ hàng rỗng")
        print("5. Tính tổng tiền")
        print("6. Hiển thị giỏ hàng")
        print("7. Xóa toàn bộ giỏ hàng")
        print("==========================")
        
        choice = input("Nhập lựa chọn của bạn (2-7): ")

        if choice == '2':
            name = input("Nhập tên sản phẩm: ")
            try:
                price = float(input("Nhập giá sản phẩm: "))
                cart.add_product(name, price)
            except ValueError:
                print("Lỗi: Giá phải là một con số.")
        
        elif choice == '3':
            name = input("Nhập tên sản phẩm cần xóa: ")
            cart.remove_product(name)
        
        elif choice == '4':
            cart.check_empty()
        
        elif choice == '5':
            cart.get_total()
        
        elif choice == '6':
            cart.show_cart()
        
        elif choice == '7':
            cart.clear_cart()
            
        else:
            print("Lựa chọn không hợp lệ, vui lòng chọn từ 2 đến 7.")

        exit_ask = input("\nCó muốn thoát khỏi chương trình không? (nhập 'yes' để thoát): ")
        if exit_ask == "yes":
            print("Chương trình kết thúc.")
            break

if __name__ == "__main__":
    main()
```

```python
#W8A6

class Rectangle:
    def __init__(self, w, h):
        self.w = w
        self.h = h

    def area(self):
        return self.w * self.h

    def perimeter(self):
        return 2 * (self.w + self.h)

    def scale(self, k):
        self.w = self.w * k
        self.h = self.h * k

if __name__ == "__main__":
    try:
        input_str = input("Nhập w, h, k: ")

        values = list(map(float, input_str.split()))
        
        if len(values) != 3:
            print("Vui lòng nhập đúng 3 số: chiều rộng, chiều cao và hệ số k.")
        else:
            w, h, k = values

            rect = Rectangle(w, h)

            rect.scale(k)

            a = rect.area()
            p = rect.perimeter()

            print(f"{a:.2f} {p:.2f}")

    except ValueError:
        print("Lỗi: Dữ liệu nhập vào phải là số thực.")
```

```python
#W8A7

import math

class Fraction:
    def __init__(self, num, den):
        if den == 0:
            raise ValueError("Mẫu số không thể bằng 0")

        if den < 0:
            num = -num
            den = -den
            
        common = math.gcd(num, den)
        self.num = num // common
        self.den = den // common

    def add(self, other):
        new_num = self.num * other.den + other.num * self.den
        new_den = self.den * other.den
        return Fraction(new_num, new_den)

    def sub(self, other):
        new_num = self.num * other.den - other.num * self.den
        new_den = self.den * other.den
        return Fraction(new_num, new_den)

    def mul(self, other):
        new_num = self.num * other.num
        new_den = self.den * other.den
        return Fraction(new_num, new_den)

    def div(self, other):
        if other.num == 0:
            raise ValueError("Không thể chia cho phân số có giá trị 0")
        new_num = self.num * other.den
        new_den = self.den * other.num
        return Fraction(new_num, new_den)

    def __str__(self):
        return f"{self.num}/{self.den}"

if __name__ == "__main__":
    try:
        line = input("Nhập biểu thức (a b op c d): ")
        parts = line.split()

        if len(parts) != 5:
            print("Lỗi: Vui lòng nhập đúng định dạng 'a b op c d' (ví dụ: 1 2 + 3 4)")
        else:
            a, b, op, c, d = parts
            a, b, c, d = int(a), int(b), int(c), int(d)
            
            f1 = Fraction(a, b)
            f2 = Fraction(c, d)

            result = None
            if op == '+':
                result = f1.add(f2)
            elif op == '-':
                result = f1.sub(f2)
            elif op == '*':
                result = f1.mul(f2)
            elif op == '/':
                result = f1.div(f2)
            else:
                print("Lỗi: Phép tính không hợp lệ (chỉ dùng +, -, *, /)")

            if result:
                print(result)

    except ValueError as e:
        print(f"Lỗi giá trị: {e}")
    except ZeroDivisionError:
        print("Lỗi: Chia cho 0.")
```

```python
#W8A8

class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = float(balance)

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount

    def __str__(self):
        return f"{self.owner}: {self.balance}"

if __name__ == "__main__":
    try:
        line1 = input("Nhập tên và số dư ban đầu: ").split()
        if len(line1) < 2:
            print("Lỗi: Nhập sai định dạng.")
        else:
            owner = line1[0]
            initial_balance = float(line1[1])

            account = BankAccount(owner, initial_balance)

            n = int(input("Nhập số lệnh: "))

            for i in range(n):
                command_line = input().split()
                cmd = command_line[0]      
                amount = float(command_line[1]) 

                if cmd == "DEPOSIT":
                    account.deposit(amount)
                elif cmd == "WITHDRAW":
                    account.withdraw(amount)
            
            print(f"{account.balance:.2f}")

    except ValueError:
        print("Lỗi: Dữ liệu nhập vào không hợp lệ.")
```

```python
#W8A10

class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = int(year) 

class Library:
    def __init__(self):
        self.books = []

    def add(self, book):
        self.books.append(book)

    def count_by_author(self, author_name):
        count = 0
        for book in self.books:
            if book.author == author_name:
                count += 1
        return count

    def find_by_year(self, year):
        count = 0
        for book in self.books:
            if book.year == year:
                count += 1
        return count

if __name__ == "__main__":
    try:
        line = input("Nhập số thao tác: ")
        if line:
            n = int(line)
            
            lib = Library()
            
            for _ in range(n):
                command_line = input().strip()

                if command_line.startswith("ADD"):
                    content = command_line[4:]
                    parts = content.split(';')
                    
                    if len(parts) == 3:
                        title = parts[0]
                        author = parts[1]
                        year = parts[2]
                        new_book = Book(title, author, year)
                        lib.add(new_book)

                elif command_line.startswith("COUNTYEAR"):
                    year_str = command_line[10:] 
                    print(lib.find_by_year(int(year_str)))

                elif command_line.startswith("COUNT"):
                    author_name = command_line[6:]
                    print(lib.count_by_author(author_name))
                    
    except ValueError:
        print("Lỗi: Dữ liệu nhập vào không hợp lệ.")
```

```python
#W8A11

import math

class Vector3D:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z

    def subtract(self, other):
        return Vector3D(self.x - other.x, self.y - other.y, self.z - other.z)

    def cross_product(self, other):
        cx = self.y * other.z - self.z * other.y
        cy = self.z * other.x - self.x * other.z
        cz = self.x * other.y - self.y * other.x
        return Vector3D(cx, cy, cz)

    def dot_product(self, other):
        return self.x * other.x + self.y * other.y + self.z * other.z

    def magnitude(self):
        return math.sqrt(self.x**2 + self.y**2 + self.z**2)

def solve():
    try:
        points = []
        for _ in range(4):
            line = input().strip()
            if not line: break
            parts = list(map(float, line.split()))
            points.append(Vector3D(parts[0], parts[1], parts[2]))

        if len(points) < 4:
            return

        A, B, C, D = points[0], points[1], points[2], points[3]
        AB = B.subtract(A)
        BC = C.subtract(B)
        CD = D.subtract(C)

        n1 = AB.cross_product(BC)

        n2 = BC.cross_product(CD)

        dot_val = n1.dot_product(n2)
        mag_n1 = n1.magnitude()
        mag_n2 = n2.magnitude()

        if mag_n1 == 0 or mag_n2 == 0:
            print("Lỗi: Các điểm thẳng hàng, không tạo thành mặt phẳng.")
        else:
            cos_phi = abs(dot_val) / (mag_n1 * mag_n2)

            cos_phi = min(1.0, max(-1.0, cos_phi))

            phi_rad = math.acos(cos_phi)
            phi_deg = math.degrees(phi_rad)

            print(f"{phi_deg:.2f}")

    except ValueError:
        print("Lỗi: Dữ liệu nhập vào phải là số thực.")
    except Exception as e:
        print(f"Đã xảy ra lỗi: {e}")

if __name__ == "__main__":
    solve()
```

```python
#W8A12

class ComplexNumber:
    def __init__(self, r, i):
        self.r = r
        self.i = i
    
    def multiply(self, other):
        real_part = self.r * other.r - self.i * other.i
        imag_part = self.r * other.i + self.i * other.r
        return ComplexNumber(real_part, imag_part)

if __name__ == "__main__":
    try:
        r1, i1 = map(float, input().split())
        r2, i2 = map(float, input().split())
        
        c1 = ComplexNumber(r1, i1)
        c2 = ComplexNumber(r2, i2)
        
        res = c1.multiply(c2)
        
        out_r = int(res.r) if res.r.is_integer() else res.r
        out_i = int(res.i) if res.i.is_integer() else res.i
        print(f"{out_r} {out_i}")
        
    except ValueError:
        print("Lỗi nhập liệu")
```

```python
#W8A13

import math

class Triangle:
    def __init__(self, edge1, edge2, edge3):
        self.edge1 = edge1
        self.edge2 = edge2
        self.edge3 = edge3

    def getArea(self):
        a, b, c = self.edge1, self.edge2, self.edge3

        if (a + b > c) and (a + c > b) and (b + c > a):
            p = (a + b + c) / 2
            area = math.sqrt(p * (p - a) * (p - b) * (p - c))
            return area
        else:
            return -1

if __name__ == "__main__":
    try:
        line = input("Nhập 3 cạnh tam giác: ")
        parts = list(map(int, line.split()))
        
        if len(parts) != 3:
            print("invalid")
        else:
            t = Triangle(parts[0], parts[1], parts[2])
            result = t.getArea()
            
            if result == -1:
                print("invalid")
            else:
                if result.is_integer():
                    print(int(result))
                else:
                    print(f"{result:.2f}")

    except ValueError:
        print("invalid")
```
